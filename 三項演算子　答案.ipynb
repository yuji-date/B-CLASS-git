```python
data1 = 10
data2 = 20

## いままで
if data1 == data2:
    i = "正しい"
else:
    i = "まちがい"
print(i)
```

```python
## 三項演算子を使う
i = "正しい" if data1 == data2 else "まちがい"
print(i)
```

三項演算子のフォーマット

(if文が成り立った時の処理 　if 　　条件式 　else 　　if文が成り立たなかった時の処理)
　　　　　　　１　　　　　　　２　　　３　　　　４　　　　　　　　　５
       

```python
# 世界のナベアツ
for i in range(1,41):
    if i % 3 == 0 or int(i/10) == 3:
        print("アホ")
    else:
        print(i)

## 三項演算子のみ
for i in range(1,41):
    print("アホ" if i % 3 == 0 or int(i/10) == 3 else i)

## リスト内包表記も使う
print( ["アホ" if i % 3 == 0 or int(i/10) == 3 else i for i in range(1,40)] )

## if文のみの関数として示す
def is_aho(num):
    if i % 3 == 0 or int(i/10) == 3:
        return "アホ"
    else:
        return i
    
for i in range(1,41):
    print( is_aho(i) )
```

```python
#Fizz Buzz を一行で作る
#まずは普通に作る
num=range(0,100)
for i in num:
    if i%15==0:
        print('FizzBuzz')
    elif i%3==0:
        print('Fizz')
    elif i%5==0:
        print('Buzz')
    else:
        print(i)
```

```python
#一行にする
print(['FizzBuzz' if i%15==0 else 'Fizz' if i%3==0 else 'Buzz' if i%5==0 else i for i in num])
```
